## Match Statistical Summary

```python
# Iterate over unique combinations of 'opposition', 'team', 'year'
df_unique=df[['opposition','team','start_date']].drop_duplicates()
for opposition, team, year in df_unique.itertuples(index=False):
    
    # Filter the DataFrame based on conditions
    data = df[(df['team']==team)& (df['start_date']==year)&(df['opposition'] == opposition) & (df['bat_or_bowl'] == 'bowl')]
    
    # Calculate batting runs
    batting_runs = data.groupby('start_date')['runs'].sum()
    runs = batting_runs.values[0]  # Assuming a single value for runs
    
    # Calculate bowl played
    bowl_played = data.groupby('start_date')['bb_bf'].sum()
    bp = bowl_played.values[0]  # Assuming a single value for bowl played
    
    # Calculate count
    total_wkts=data.groupby('start_date')['wkts'].sum()
    wkts_sum=total_wkts.values[0]
    wkts_sum=np.abs(wkts_sum-10)
    # Print information for each match separately
    print(f"{opposition} made {runs} runs in {bp} balls by {wkts_sum} wickets on date {batting_runs.index[0].strftime('%Y-%m-%d')}")
```

# Team Economy Rate

In the context of the ICC Cricket World Cup, the team economy rate refers to the average number of runs conceded by a team’s bowlers per over. It’s a crucial metric that reflects how economically a team’s bowlers are able to contain the opposition’s batting lineup.

Here’s how it works:

Economy Rate (ER) is calculated by dividing the total number of runs conceded by the total number of overs bowled.
A lower economy rate indicates that the bowlers are efficient in restricting runs, while a higher rate suggests that they are leaking runs more freely.
For instance, during the ICC Cricket World Cup 2023/24, some bowlers maintained exceptional economy rates, contributing significantly to their team’s success1. It’s an essential statistic for assessing a team’s overall bowling performance and strategizing during matches.

Remember, a tight economy rate can put pressure on the opposition batsmen and influence the course of a game!
In most circumstances, the lower the economy rate is, the better the bowler is performing.

### Best Bowling Economy

```python
plt.figure(figsize=(10,3))
sns.set(style='ticks')
custom_color = ['red', 'blue', 'green', 'yellow', 'orange', 'purple', 'teal']

total_runs = df.groupby('team')['runs'].sum()
total_wickets = df.groupby('team')['wkts'].sum()
economy_rate = df.groupby('team')['econ'].mean()

team_summary = pd.DataFrame({
    'Total Runs': total_runs,
    'Total Wickets': total_wickets,
    'Average Economy Rate': economy_rate
})

sorted_teams = team_summary.sort_values(by='Average Economy Rate', ascending=True)

sorted_teams['Average Economy Rate'].plot(kind='bar', color=custom_color)
plt.xlabel('Team')
plt.ylabel('Economy')
plt.title('Team-wise Economy Rate')
plt.show()

best_teams = sorted_teams.head(10)
print(best_teams)
```

### Overall Team Contribution

```python
team = df[df['team']=='AUS']

total_runs = team['runs'].sum()
total_wkts = team['wkts'].sum()

player_runs_contribution = (team.groupby('player')['runs'].sum() /total_runs) * 100
player_wkts_contribution = (team.groupby('player')['wkts'].sum()/total_wkts ) * 100
player_contributions = pd.concat([player_runs_contribution, player_wkts_contribution], axis=1)
player_contributions.columns = ['Runs','Wickets']  # Assigning customized col name

player_contributions['Combined'] = player_contributions['Runs'] + player_contributions['Wickets']
player_contributions = player_contributions.sort_values('Combined', ascending=False)

# Custom colors for bars
colors = ['#1b9e77', '#a9f971', '#fdaa48', '#6890F0', '#A890F0']

plt.figure(figsize=(8, 6))
plt.bar(player_contributions.index, player_contributions['Combined'], color=colors)
plt.xlabel('Player')
plt.ylabel('Contribution Percentage')
plt.title('Overall Player Contribution in Runs and Wickets')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
